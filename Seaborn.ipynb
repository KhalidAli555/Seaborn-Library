# **Bar Graph in Seaborn Library**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df2=sns.load_dataset("penguins")
df2
```

```python
sns.barplot(x=df2.island,y=df2.bill_length_mm,color="green")
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df2,hue="sex") # hue=sex means data divide on the base of gender
plt.title("Bar Plot")
plt.show()
```

# **Order Parameter**

```python
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=df2,hue="sex",order=order1) # order gives you x-asis sequentially(orderly)
plt.show()
```

# **Hue Order**

```python
order1=["Dream","Torgersen","Biscoe"]                                                               # palette give specific color to male and female
sns.barplot(x="island",y="bill_length_mm",data=df2,hue="sex",order=order1,hue_order=['Female','Male'],palette="prism") # hue order change the sex means first female and second Male represent in graph
plt.show()
```

# **Saturation**(saturation give visibility to graph its value is 0 to 1)

```python
sns.barplot(x="island",y="bill_length_mm",data=df2,saturation=0.2,color="green")  # saturation gives visibility to graph
plt.show()
```

# **histogram Graph**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df2["flipper_length_mm"])
plt.title("Histogram")
plt.show()
```

# **Bins Parameter in Histogram(Give specific range in graph)**

```python
sns.histplot(df2["flipper_length_mm"],bins=[170,180,190,200,210,220,230])
plt.title("Histogram")
plt.show()
```

# **kde parameter(to show density of graph)**

```python
sns.histplot(df2["flipper_length_mm"],bins=[170,180,190,200,210,220,230],kde=True) # kde(the line in a graph is kde)
plt.title("Histogram")
plt.show()
```
